```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data=pd.read_csv('budget.csv')
data
```

```python
plt.scatter(data['Year'],data['Budget'],color='red',marker='o')
slope,intercept=np.polyfit(data['Year'],data['Budget'],1)
x=np.linspace(data['Year'].min(),data['Year'].max(),100)
y=slope*x+intercept
plt.plot(x,y,color='blue')
plt.title('Pakistan Budget 2015-2024')
plt.xlabel('Years')
plt.ylabel('Budget')
plt.show()
```

```python
new_data=data.drop('Budget',axis='columns')
new_data
```

```python
model=linear_model.LinearRegression()
model
```

```python
model=model.fit(new_data,data.Budget)
model
```

```python
model.predict([[2025]])
```
